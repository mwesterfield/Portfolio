# Analysis of time-to-resolve Get It Done San Diego requests
## Notebook 2: Visualization
### How do San Diegans communicate with their City government?
This notebook highlights one particular aspect of the publicly-available “Get It Done San Diego” dataset. One feature tracked in the data (‘case_origin’) is how the customer made the request. The goal of this sub-analysis is to provide a description of how San Diego residents communicate with the City government, and determine whether there are any district-specific differences.

#### Load all necessary libraries

```python
import pandas as pd
from datetime import timedelta
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
notebook_path = os.path.abspath("Notebook2-Visualization.ipynb")
```

#### Import data/gid_with_demo.csv

```python
filename_csv = os.path.join(os.path.dirname(notebook_path), "data/gid_with_demo.csv")
gid_with_demo = pd.read_csv(filename_csv, index_col=0, parse_dates=['requested_datetime','updated_datetime'],)

# We are only interested in requests from the public
gid_public = gid_with_demo.copy()
gid_public = gid_public[(gid_public.case_origin == 'Mobile') |
                            (gid_public.case_origin == 'Web') |
                            (gid_public.case_origin == 'Phone') |
                            (gid_public.case_origin == 'Email') |
                            (gid_public.case_origin == 'Walk-In') |
                            (gid_public.case_origin == 'Letter')]
```

### Communication preferences over San Diego as a whole
Overall, most requests are submitted via the mobile phone app (153,936 or 52.1%) or online through the City website (102,109 or 34.6%). A sizeable number of requests are made over the phone (35,282 or 11.9%), and a much smaller number are made by email (3,967 or 1.3%). Fewer than 20 requests were made by letter or by walk-in combined (< 1%), see Figure 1, below.

```python
# Plot number of requests by category
gid_public.groupby('case_origin').service_request_id.size().sort_values(ascending=False).plot(kind='bar', 
                                                                                            color=(0.2, 0.4, 0.6, 0.6))
plt.xticks(rotation=45)
plt.xlabel('Contact method')
plt.ylabel('Number of requests')
plt.title('Fig 1 - Request frequency by communication method')
plt.tight_layout()
plt.show()
#plt.savefig('CaseOrigin_Totals.png')
```

By tracking the number of requests made over time since the app’s introduction in May 2016, we can see that mobile app and website usage have been increasing steadily, whereas requests made by phone have not made quite the same gain (see Figure 2, below). Email requests, in contrast, have declined over time.

```python
# First, create df with index requested_datetime (so we can create running averages)
#gid_public['requested_datetime'] = pd.to_datetime(gid_public.requested_datetime)
gid_public_bytime = gid_public.copy()
gid_public_bytime = gid_public_bytime.set_index('requested_datetime')

# Second, resample dataframe created above by month and calculate
# total requests grouped by communication method (case_origin)
monthly_summary_all = pd.DataFrame()
monthly_summary_all = gid_public_bytime.groupby('case_origin').service_request_id.resample('M').size()

# Time series plot of request totals by communication type
monthly_summary_all.unstack(0).plot(color=['#d62728','#8c564b','#1f77b4','#ff7f0e','#9467bd','#2ca02c'])
plt.xlabel('')
plt.ylabel('Number of requests')
plt.title('Fig 2 - Monthly requests according to contact method')
plt.tight_layout()
plt.show()
#plt.savefig('CaseOrigin_Monthly.png')
```

### Communication preference broken down by City District
When service requests are further categorized by district, it is evident that the mobile app is the preferred communication method for half of San Diego’s districts (e.g., Districts 2, 3, 4, 8, and 9), web access is preferred in Districts 1 and 5, and Districts 6 and 7 are more balanced between mobile and web access (see Figure 3, below).

```python
# Extract number of requests from only mobile, phone, 
# and web requests grouped by district
temp_summary = pd.DataFrame()
temp_summary = gid_public[(gid_public.case_origin=='Mobile') | (gid_public.case_origin=='Phone') |
                         (gid_public.case_origin=='Web')].groupby('district').case_origin.value_counts()
temp_summary = temp_summary.unstack(1)

# Calculate ratio of mobile, phone, and web requests 
# from the sum of those requests only
temp_summary.loc[:,"Mobile":"Web"] = temp_summary.loc[:,"Mobile":"Web"].div(temp_summary.sum(axis=1), axis=0)

# Plot ratio distribution for each district 
temp_summary.plot.barh(xlim=[0,1], stacked=True)
plt.legend(bbox_to_anchor=(0.0, 1.02, 1.0, 0.102), loc=3,
       ncol=3, mode="expand", borderaxespad=0.0)
plt.xlabel('Proportion of Requests')
plt.ylabel('District')
plt.axvline(x=0.5, color='k')
plt.title('Fig 3 - Ratio of Mobile, Phone, and Web requests per district', y=1.12, fontsize=14)
plt.tight_layout()
plt.show()
#plt.savefig('CaseOrigin_ByDistrict.png')
```

By combining demographic information (obtained from the 2010 Census) per zipcode with mobile app, website, or phone usage rate, it appears that both household income and population density strongly predict which communication method will be more popular within a zipcode. More dense, lower income neighborhoods are more likely to use the mobile app, whereas less dense, more wealthy neighborhoods are more likely to use the City website or to call the City directly.

```python
# First, extract proportion of mobile, web and phone requests
# grouped by district
temp_summary = pd.DataFrame()
temp_summary = gid_public[(gid_public.case_origin=='Mobile') |
                         (gid_public.case_origin=='Phone') |
                         (gid_public.case_origin=='Web')].groupby('zipcode').case_origin.value_counts()
temp_summary = temp_summary.unstack(1)
temp_summary.loc[:,"Mobile":"Web"] = temp_summary.loc[:,"Mobile":"Web"].div(temp_summary.sum(axis=1), axis=0)

# Second, add columns for household income and age, averaged over district
temp_summary['household_income'] = gid_public.groupby('zipcode').household_income.mean()
temp_summary['pop_density'] = gid_public.groupby('zipcode').pop_density.mean()

# Scatterplots
f, axes = plt.subplots(2, 3, sharey=True)
axes[0, 0].scatter(temp_summary.household_income, temp_summary.Mobile)
axes[0, 1].scatter(temp_summary.household_income, temp_summary.Web)
axes[0, 2].scatter(temp_summary.household_income, temp_summary.Phone)
axes[1, 0].scatter(temp_summary.pop_density, temp_summary.Mobile)
axes[1, 1].scatter(temp_summary.pop_density, temp_summary.Web)
axes[1, 2].scatter(temp_summary.pop_density, temp_summary.Phone)

for i, row in enumerate(axes):
    for j, cell in enumerate(row):
        if (j==0):
            cell.set_ylabel('Proportion of Requests')
        if (j==1) & (i==0):
            cell.set_xlabel('Median income ($)')
        if (j==1) & (i==1):
            cell.set_xlabel('Population density')
        if (i==0) & (j==0):
            cell.set_title('Mobile')
        if (i==0) & (j==1):
            cell.set_title('Web')
        if (i==0) & (j==2):
            cell.set_title('Phone')

f.suptitle('Fig 4 - Communication method may be correlated with\n income and population density across zipcodes', fontsize=12, y=1.08)
plt.tight_layout()
plt.show()
#plt.savefig('DemoVsCaseOrigin_Scatter.png')
```

### Effectiveness of different communication methods
The percentage of service requests that have been resolved and the amount of time it takes to close a service request are shown in Figures 5 and 6. 

There are two types of 'Closed' requests in the database: 'Closed' and 'Closed - Referred.' Unfortunately the online data dictionary does not describe the difference between these two outcomes (if any). If 'Closure' of a request is the primary measure of effectiveness, it appears that residents are most successful when they submit requests over email or phone, whereas requests submitted via the mobile app and website are least effective (see Figure 5, left column). If both 'Closed' and 'Closed - Referred' requests are considered, mobile and web submissions are the most effective (Fig.5, right column). While requests made by letter are 100% successful, this is not meaningful as there were less than 20 such requests.

```python
# Left plot - CLOSED requests grouped by communication method
# First create a new df with all closed requests
gid_closed = gid_public[gid_public.status=='Closed']

# Second, divide number of closed requests per case_origin category
# by original number of requests per case_origin categories, then plot
plt.figure(1, figsize=(10,3))
plt.subplot(1,2,1)
(gid_closed[(gid_closed.case_origin=='Email') |
            (gid_closed.case_origin=='Letter') |
            (gid_closed.case_origin=='Mobile') |
            (gid_closed.case_origin=='Phone') |
            (gid_closed.case_origin=='Walk-In') |
            (gid_closed.case_origin=='Web')].groupby('case_origin').size() /
 gid_public[(gid_public.case_origin=='Email') |
            (gid_public.case_origin=='Letter') |
            (gid_public.case_origin=='Mobile') |
            (gid_public.case_origin=='Phone') |
            (gid_public.case_origin=='Walk-In') |
            (gid_public.case_origin=='Web')].groupby('case_origin').size()).plot(kind='bar',
                                                                              color=('#d62728','#8c564b','#1f77b4','#ff7f0e','#9467bd','#2ca02c'),
                                                                              ylim=[0,1])
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Closed Requests\n(% of total)')
plt.title('Closed Requests Only',fontsize=12)

# Right plot - CLOSED+REFERRED requests grouped by communication method
# First create a new df with all closed requests
gid_resolved = gid_public[gid_public.resolved==1]

# Second, divide number of closed requests per case_origin category
# by original number of requests per case_origin categories, then plot
plt.subplot(1,2,2)
(gid_resolved[(gid_resolved.case_origin=='Email') |
            (gid_resolved.case_origin=='Letter') |
            (gid_resolved.case_origin=='Mobile') |
            (gid_resolved.case_origin=='Phone') |
            (gid_resolved.case_origin=='Walk-In') |
            (gid_resolved.case_origin=='Web')].groupby('case_origin').size() /
 gid_public[(gid_public.case_origin=='Email') |
            (gid_public.case_origin=='Letter') |
            (gid_public.case_origin=='Mobile') |
            (gid_public.case_origin=='Phone') |
            (gid_public.case_origin=='Walk-In') |
            (gid_public.case_origin=='Web')].groupby('case_origin').size()).plot(kind='bar',
                                                                              color=('#d62728','#8c564b','#1f77b4','#ff7f0e','#9467bd','#2ca02c'),
                                                                              ylim=[0,1])
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Closed + Referred\n(% of total)')

plt.subplots_adjust(top=0.8, wspace=0.5)
plt.title('Closed + Referred Requests',fontsize=12)
plt.suptitle('Fig. 5 - Outcome According to Communication Method', fontsize=14)
plt.show()
#plt.tight_layout
#plt.savefig('ResolvedRequestsPerCommMethod.png')
```

If only closed requests are considered, communication via mobile app, phone, and website have the fastest closure times, with a median closure times of approximately 10 days (see Figure 6). \[Note the large number of outliers in those categories, however; these outliers will ultimately constrain the types of predictive models likely to be successful with this data\]

```python
sns.catplot(x="case_origin", y="days_till_closed", kind="box", 
            data=gid_closed)
plt.xticks(rotation=45)
plt.xlabel('')
plt.ylabel('Number of days until request is closed')
plt.title('Fig 6 - Effectiveness\n(According to Time-to-close)',fontsize=14)
plt.tight_layout()
plt.show()
#plt.savefig('DaysVsCase_origin.png')
```

### Conclusion
The City of San Diego appears to have been successful at transitioning San Diego residents to this new method of communicating with city departments, as seen by the steady increase over time in requests submitted both over the mobile app and the City website. There is a definite district-specific difference in communication. It appears that districts with wealthier and older residents prefer to use web- or phone-based means of communication, whereas districts with less-wealthy and younger residents are far more likely to use the mobile app.

Of the three most popular ways that residents communicate with the City, it appears that mobile app and website use are the most effective in that requests submitted by those two methods have the highest resolution rate. The amount of time it takes to close a request submitted by mobile app, over the phone, or through the website is relatively quick (~ 10 days) and approximately the same for each method.
